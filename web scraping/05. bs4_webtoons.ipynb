```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://comic.naver.com/webtoon/weekday"
res = requests.get(url)
res.raise_for_status()
```

```python
soup = BeautifulSoup(res.text, "lxml")
```

### 네이버 웹툰 전체 목록 가져오기

```python
# class 속성이 title인 모든 "a" element를 반환
cartoons = soup.find_all("a", attrs={"class":"title"})
```

```python
# 웹툰 홈페이지에 나와 있는 모든 웹툰의 제목 출력하기
for cartoon in cartoons:
    print(cartoon.get_text())
```

```python
# 조조 코믹스 정보 가져오기
url = "https://comic.naver.com/webtoon/list?titleId=774862"
res = requests.get(url)
res.raise_for_status()
```

```python
soup = BeautifulSoup(res.text, "lxml")
```

```python
cartoons = soup.find_all("td", attrs = {"class":"title"})
```

```python
title = cartoons[0].a.get_text()
print(title)
```

```python
# 웹툰 제목과 링크 출력하기
link = cartoons[0].a["href"]

print(title)
print("https://comic.naver.com" + link)
```

```python
for cartoon in cartoons:
    title = cartoon.a.get_text()
    link = "https://comic.naver.com" + cartoon.a["href"] + "\n"
    
    print(title, link)
```

## 웹툰의 평점가져오기

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://comic.naver.com/webtoon/list?titleId=774862&weekday"
res = requests.get(url)
res.raise_for_status()
```

```python
soup = BeautifulSoup(res.text, "lxml")
```

```python
recode = soup.find_all("div", attrs={"class":"rating_type"})
```

```python
for re in recode:
    rate = re.find("strong").get_text()
    print(rate)
```

```python
total_rates = 0

for re in recode:
    rate = re.find("strong").get_text()
    print(rate)
    total_rates += float(rate)
print("전체 점수 : ", total_rates)
print("평균 점수 : ", total_rates / len(recode))
```
