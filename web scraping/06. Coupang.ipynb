```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
url = "https://www.coupang.com/np/search?q=%EB%85%B8%ED%8A%B8%EB%B6%81&channel=recent&component=&eventCategory=SRP&trcid=&traid=&sorter=scoreDesc&minPrice=&maxPrice=&priceRange=&filterType=&listSize=36&filter=&isPriceRange=false&brand=&offerCondition=&rating=0&page=1&rocketAll=false&searchIndexingToken=1=5&backgroundColor="
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"}
res = requests.get(url, headers = headers)
res.raise_for_status()
```

```python
soup = BeautifulSoup(res.text, "lxml")
```

```python
#  class 속성이 search-product로 시작하는 모든 "li" element를 반환
items = soup.find_all("li", attrs = {"class" : re.compile("search-product")})
```

```python
# 상품 이름 가져오기
print(items[0].find("div", attrs = {"class":"name"}).get_text())
```

```python
# 가격 가져오기
items[0].find("strong", attrs = {"class":"price-value"}).get_text()
```

```python
# 평점 가져오기
items[0].find("em", attrs = {"class":"rating"}).get_text()
```

```python
# 리뷰 수 가져오기
items[0].find("span", attrs = {"class":"rating-total-count"}).get_text()
```

```python
for item in items:
    name = item.find("div", attrs = {"class":"name"}).get_text()
    price = item.find("strong", attrs = {"class":"price-value"}).get_text()
    rate = item.find("em", attrs = {"class":"rating"}).get_text()
    rate_cnt = item.find("span", attrs = {"class":"rating-total-count"}).get_text()
    link = "http://coupang.com" + item.a["href"]
    
    #print(name, price, rate, rate_cnt, link)
    print(f"상품명 : {name}")
    print(f"가격 : {price}")
    print(f"평점 : {rate}")
    print(f"리뷰 수 : {rate_cnt}")
    print(f"링크 : {link} \n")
```

```python
# 리뷰 50개 이상, 평점 4.5 이상 되는 것만 조회
for item in items:
    ad_badge = item.find("span", attrs = {"class":"ad-badge-text"})
    if ad_badge:
        print("< 광고 상품은 제외합니다 >")
        continue
        
    name = item.find("div", attrs = {"class":"name"}).get_text()
    price = item.find("strong", attrs = {"class":"price-value"}).get_text() + "원"

    rate = item.find("em", attrs = {"class":"rating"})
    if rate:
        rate = rate.get_text()
    else:
        print("< 평점 없는 상품은 제외합니다 >")
        continue
        
    rate_cnt = item.find("span", attrs = {"class":"rating-total-count"})
    if rate_cnt:
        rate_cnt = rate_cnt.get_text()
        rate_cnt = rate_cnt[1:-1]
    else:
        print("< 평점 수 없는 상품은 제외합니다 >")
        continue
        
    if float(rate) >= 4.5 and int(rate_cnt) >= 50:
        print(name, price, rate, rate_cnt)
```
